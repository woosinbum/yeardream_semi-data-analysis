```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.options.display.float_format = '{:.0f}'.format
```

```python
from pyspark.sql import SparkSession
import databricks.koalas as ks

spark = SparkSession.builder.getOrCreate()
```

```python
def read_files_spark(local_dir,header):
    df = spark.read.csv(local_dir,encoding='cp949',header=header)
    return df
```

```python
sales_df = read_files_spark('./상권-추정매출/서울시우리마을가게상권분석서비스(상권-추정매출)_2020.csv',True).to_koalas()
```

```python
income_df = read_files_spark('./상권배후지-소득소비',True).to_koalas()
```

```python
sales_df = sales_df[sales_df['기준_년_코드'] == '2020']
```

```python
sales_df = sales_df.groupby('상권_코드').agg('sum')
```

```python
income_df = income_df[income_df['기준 년 코드'] == '2020']
```

```python
income_df = income_df.groupby('상권_코드').agg('sum')
```

```python
sales_df = sales_df[['분기당_매출_금액']].rename(columns={'분기당_매출_금액':'매출'})
```

```python
sales_df
```

```python
income_df['월_평균_소득_금액'] /= 4
```

```python
income_df = income_df.drop(columns=['기준 년 코드','기준_분기_코드','상권_구분_코드','상권_구분_코드_명','상권_코드_명'])
```

```python
income_df
```

```python
df = sales_df.merge(income_df,left_index=True,right_index=True)
```

```python
res = []
for i in df.columns[1:]:
    res.append((df['매출'].corr(df[i]),i))
```

```python
res.sort(reverse=True)
res
```
