```python
def csv_into_df(year):
    if year < 2020:
        csv_title = '상권-추정매출/서울시 우리마을가게 상권분석서비스(상권-추정매출)_' + str(year) + '.csv'
    else:
        csv_title = '상권-추정매출/서울시우리마을가게상권분석서비스(상권-추정매출)_' + str(year) + '.csv'
    obj = s3_client.get_object(Bucket="elice-team5-bucket", Key=csv_title)
    df = pd.read_csv(io.BytesIO(obj["Body"].read()), encoding='cp949', low_memory=False)
    return df
```

# csv 파일 concat 으로 df에 넣기

```python
df = pd.DataFrame()
for year in range(2016, 2021):
    df = pd.concat([df, csv_into_df(year)])
df
```

# '기준_년_코드', '상권_코드' 그룹화

```python
grouped_df = df.groupby(['기준_년_코드','상권_코드']).sum()
grouped_df
```

# 필요한 데이터 index 만들기

```python
grouped_df['월별_평균_매출'] = grouped_df['분기당_매출_금액']/grouped_df['점포수']/4
grouped_df['전년도비_매출_증감'] = np.nan
grouped_df['전년도비_매출_증감률'] = np.nan
grouped_df['전년도비_점포수_증감'] = np.nan
grouped_df['전년도비_점포수_증감률'] = np.nan
```

# 전년도비_매출_증감

```python
for year in range(2017, 2021):
    grouped_df['전년도비_매출_증감'].loc[year].loc[grouped_df.loc[year].index.isin(grouped_df.loc[year-1].index)] = grouped_df.loc[year, '월별_평균_매출'].filter(items=grouped_df.loc[year-1].index, axis=0) - grouped_df.loc[year-1, '월별_평균_매출'].filter(items=grouped_df.loc[year].index, axis=0)
grouped_df.loc[2018]
```

# 전년도비_매출_증감률

```python
for year in range(2017, 2021):
    grouped_df['전년도비_매출_증감률'].loc[year].loc[grouped_df.loc[year].index.isin(grouped_df.loc[year-1].index)] = (grouped_df.loc[year, '월별_평균_매출'].filter(items=grouped_df.loc[year-1].index, axis=0) / grouped_df.loc[year-1, '월별_평균_매출'].filter(items=grouped_df.loc[year].index, axis=0)).values * 100 - 100
grouped_df.loc[2019]
```

# 전년도비_점포수_증감

```python
for year in range(2017, 2021):
    grouped_df['전년도비_점포수_증감'].loc[year].loc[grouped_df.loc[year].index.isin(grouped_df.loc[year-1].index)] = grouped_df.loc[year, '점포수'].filter(items=grouped_df.loc[year-1].index, axis=0) - grouped_df.loc[year-1, '점포수'].filter(items=grouped_df.loc[year].index, axis=0)
grouped_df.loc[2018]
```

# 전년도비_점포수_증감률

```python
for year in range(2017, 2021):
    grouped_df['전년도비_점포수_증감률'].loc[year].loc[grouped_df.loc[year].index.isin(grouped_df.loc[year-1].index)] = (grouped_df.loc[year, '점포수'].filter(items=grouped_df.loc[year-1].index, axis=0) / grouped_df.loc[year-1, '점포수'].filter(items=grouped_df.loc[year].index, axis=0)).values * 100 - 100
grouped_df.loc[2019]
```

# pivot_table 생성

```python
pivot_df = pd.pivot_table(grouped_df, index=['기준_년_코드', '상권_코드'], values=['분기당_매출_금액', '점포수', '월별_평균_매출', '전년도비_매출_증감', '전년도비_매출_증감률', '전년도비_점포수_증감', '전년도비_점포수_증감률'])
# 연도별 필터링 및 정렬
pivot_df.loc[2019].sort_values(by='상권_코드', ascending=True)
```
